# Getting started

This tutorial coverst the basics of computing 2D GPs with ``exoplanet`` using the classes provided by ``SpecGP``.

``exoplanet`` and ``specgp`` are primarily intended for use with ``pymc3`` as a result, all variables are actually nodes in a computational graph in the form of ``theano`` tensors. This means that in order to compute the actual value of a variable outside of the ``pymc3`` model context, we'll need to periodically call ``.eval()`` in our code.

To start, let's define a 1D ``celerite`` term:

```python
import numpy as np
import exoplanet as xo

term = xo.gp.terms.SHOTerm(log_S0=0.0, log_w0=1.0, log_Q=-np.log(np.sqrt(2)))
```

Here we've chosen the simple harmonic oscillator (SHO) term, which is a commonly used model for stellar variability. The choice is somewhat arbitrary; ``specgp`` works with any ``celerite`` term.

If we're only working in one dimension we can define the GP with this kernel alone:

```python
t = np.linspace(0, 10, 1000)
gp = xo.gp.GP(
                x=t, 
                kernel=term, 
                diag=0.01 * np.ones_like(t), 
                mean=xo.gp.means.Constant(0.0), 
                J=2
              )
```

Let's take a look at a realization of this GP:

Now, say we want to model two correlated processes in, both of which have this same time-covariance, but for which the amplitude of the variability differs by a factor of two between the processes. We can do this with ``spegp``. 

The ``KroneckerTerm`` we imported from ``specgp`` can be used to combine a ``celerite`` term with an array of scale factors as follows:

```python
from SpecGP.terms import KroneckerTerm

# scaling factor for each process
alpha = [1, 2]
kernel = KroneckerTerm(term, alpha)
```

We have to be a bit careful defining the GP now because ``diag`` and ``mean`` both need to be two dimensional. For ``diag`` we simply define a 2D array with each row containing the white noise variances for one of the correlated processes:

```python
# the white noise components for each process
# here we set the second process to have 100 times the white noise 
# of the first.
diag = np.array([0.001, 0.1])
diag = diag[:, None] * np.ones_like(t)
print(diag)
```

For the the GP mean we need to use the provided ``KroneckerConstant`` mean function which is compatible with the ``KroneckerTerm`` kernel. ``KroneckerConstant`` defines a constant mean function. We give it two values, each being the (constant) mean of one of the processes:

```python
from SpecGP.means import KroneckerConstant

mu = KroneckerConstant([0, 0])
```

Finally, we can define the 2D GP:

```python
gp = xo.gp.GP(x=t, kernel=kernel, diag=diag, mean=mu, J=2)
```

Let's take a look at a realization of this GP:

```python
n = np.random.randn(2*len(t), 1)
y = gp.dot_l(n).eval()

pl.plot(t, y[::2], '.', color='#FE4365')
pl.plot(t, y[1::2], '.', color='#3F9778')
pl.xlabel("time")
pl.ylabel("y")
```

A quick note on the structure of the vector returned by ``dot_l``: $y$ is a 1D vector with the first 2 elements consisting of the first observation for each of the two processes, the next two being the second observation at each wavelength, and so forth. We could also write $y = vec(Y)$ where $Y$ is a matrix of size $2\times N$ with $N$ the number of times and with each of the rows containing the timeseries of one of the correlated processes. For $M$ correlated processes the first $M$ elements of $y$ are the observation at the first time for each process, and so forth:  

```python
nprocesses = 10

alpha = np.linspace(1, 10, nprocesses)
kernel = KroneckerTerm(term, alpha)

diag = np.array([0.01] * nprocesses)
diag = diag[:, None] * np.ones_like(t)

mu = KroneckerConstant([0] * nprocesses)
gp = xo.gp.GP(x=t, kernel=kernel, diag=diag, mean=mu, J=2)

n = np.random.randn(nprocesses*len(t), 1)
y = gp.dot_l(n).eval()
```

```python
for i in range(nprocesses):
    pl.plot(t, y[i::nprocesses], '-', color='k')
pl.xlabel("time")
pl.ylabel("y")
```
